- –ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ H‚ÇÄ: –°—Ä–µ–¥–Ω—è—è –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ —á–∞—Å–∞—Ö –≤ –¥–≤—É—Ö –≥—Ä—É–ø–ø–∞—Ö (–ú–æ—Å–∫–≤–∞ –∏ –°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥) –Ω–µ —Ä–∞–∑–ª–∏—á–∞–µ—Ç—Å—è.

- –ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ H‚ÇÅ: –°—Ä–µ–¥–Ω—è—è –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥–µ –±–æ–ª—å—à–µ, –∏ —ç—Ç–æ —Ä–∞–∑–ª–∏—á–∏–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–æ.

    <h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ <a class="tocSkip"> </h2>
    
<b>–ù–µ–∫–æ—Ç–æ—Ä—ã–µ –∑–∞–º–µ—á–∞–Ω–∏—è –∏ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏‚ö†Ô∏è:</b> –ü—Ä–∏ —Ä–∞–±–æ—Ç–µ —Å –æ–¥–æ—Å—Ç–æ—Ä–æ–Ω–Ω–µ–π –ø—Ä–æ–≤–µ—Ä–∫–æ–π –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –ª—É—á—à–µ —Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞—Ç—å –æ—Ç –æ–±—Ä–∞—Ç–Ω–æ–≥–æ –∫ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–æ–π: –µ—Å–ª–∏ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è - `–±–æ–ª—å—à–µ`, –Ω—É–ª–µ–≤–∞—è –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å `–Ω–µ –±–æ–ª—å—à–µ`.

```python
from scipy.stats import ttest_ind
```

    <h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ <a class="tocSkip"> </h2>
    
<b>–ù–µ–∫–æ—Ç–æ—Ä—ã–µ –∑–∞–º–µ—á–∞–Ω–∏—è –∏ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏‚ö†Ô∏è:</b>
–í—Å–µ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ —Å—Ç–æ–∏—Ç –∏–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å –≤ –Ω–∞—á–∞–ª–µ –ø—Ä–æ–µ–∫—Ç–∞ - —Ç–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º –ø–æ–ª—É—á–∞—Ç–µ–ª—å —Ç–≤–æ–µ–≥–æ –æ—Ç—á—ë—Ç–∞ —Å–º–æ–∂–µ—Ç —É–∑–Ω–∞—Ç—å –æ –ø—Ä–æ–±–ª–µ–º–∞—Ö —Å –µ–≥–æ –æ–∫—Ä—É–∂–µ–Ω–∏–µ–º —Å—Ä–∞–∑—É –ø–µ—Ä–µ–¥ —Ä–∞–±–æ—Ç–æ–π —Å –ø—Ä–æ–µ–∫—Ç–æ–º, –∞ –Ω–µ –ø–æ—Å—Ä–µ–¥–∏ –∏–ª–∏ –≤ –∫–æ–Ω—Ü–µ –ø—Ä–æ–µ–∫—Ç–∞.

```python
group_2 = books[books['city']=='–ú–æ—Å–∫–≤–∞']['hours']
group_1 = books[books['city']=='–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥']['hours']

test_result = ttest_ind (
    group_1, 
    group_2, 
    equal_var = False,
    alternative = 'greater'
)
```

</div>

```python
alpha = 0.05
pv = test_result.pvalue

print(f'test_result.pvalue = {pv}')

if test_result.pvalue<alpha:
    print('–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –æ —Ç–æ–º, —á—Ç–æ —Å—Ä–µ–¥–Ω—è—è –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ —á–∞—Å–∞—Ö –æ–¥–∏–Ω–∞–∫–æ–≤–∞ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è. –°—Ä–µ–¥–Ω—è—è –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥–µ –±–æ–ª—å—à–µ.')
else:
    print('–ù—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, —á—Ç–æ —Å—Ä–µ–¥–Ω—è—è –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ —á–∞—Å–∞—Ö –æ–¥–∏–Ω–∞–∫–æ–≤–∞ –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å.')
```

–û—Ü–µ–Ω–∏—Ç–µ –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ—Å—Ç—å –≤—ã–±–æ—Ä–∫–∏ –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã—Ö —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ A/B-—Ç–µ—Å—Ç–∞. –ó–∞–¥–∞–Ω–Ω—ã–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã:

- –±–∞–∑–æ–≤—ã–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ ‚Äî 30%,

- –º–æ—â–Ω–æ—Å—Ç—å —Ç–µ—Å—Ç–∞ ‚Äî 80%,

- –¥–æ—Å—Ç–æ–≤–µ—Ä–Ω–æ—Å—Ç—å —Ç–µ—Å—Ç–∞ ‚Äî 95%.

```python
from statsmodels.stats.power import NormalIndPower
from statsmodels.stats.proportion import proportion_effectsize

# –ó–∞–¥–∞—ë–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã
alpha = 0.05  # –£—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
beta = 0.2 # –û—à–∏–±–∫–∞ –≤—Ç–æ—Ä–æ–≥–æ —Ä–æ–¥–∞, —á–∞—Å—Ç–æ 1 - –º–æ—â–Ω–æ—Å—Ç—å
power = 0.8  # –ú–æ—â–Ω–æ—Å—Ç—å —Ç–µ—Å—Ç–∞
p = 0.3 # –ë–∞–∑–æ–≤—ã–π —É—Ä–æ–≤–µ–Ω—å –¥–æ–ª–∏
mde = 0.03  # –ú–∏–Ω–∏–º–∞–ª—å–Ω—ã–π –¥–µ—Ç–µ–∫—Ç–∏—Ä—É–µ–º—ã–π —ç—Ñ—Ñ–µ–∫—Ç
```

<b>–ù–∞ –¥–æ—Ä–∞–±–æ—Ç–∫—É‚ùå:</b> `mde` —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç `0.03` - —É–º–Ω–æ–∂–∞—Ç—å –Ω–∞ `p` –Ω–µ –Ω—É–∂–Ω–æ.

<div class="alert alert-info">
<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞ <a class="tocSkip"> </h2>

–°–ø–∞—Å–∏–±–æ –±–æ–ª—å—à–æ–µ –∑–∞ –æ—Ç–≤–µ—Ç! –ò—Å–ø—Ä–∞–≤–∏–ª–∞.
</div>

```python
effect_size = proportion_effectsize(p, p + mde)
```

```python
# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –∫–ª–∞—Å—Å NormalIndPower
power_analysis = NormalIndPower()

# –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º —Ä–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏
sample_size = power_analysis.solve_power(
    effect_size = effect_size,
    power = power,
    alpha = alpha,
    ratio = 1 # –†–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤—ã–±–æ—Ä–æ–∫
)

print(f"–ù–µ–æ–±—Ö–æ–¥–∏–º—ã–π —Ä–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏ –¥–ª—è –∫–∞–∂–¥–æ–π –≥—Ä—É–ø–ø—ã: {int(sample_size)}")
```

–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (–ù1): –í —Ç–µ—á–µ–Ω–∏–µ —Å–µ–º–∏ –¥–Ω–µ–π –ø–æ—Å–ª–µ —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏ –≤ —Å–∏—Å—Ç–µ–º–µ –∫–æ–Ω–≤–µ—Ä—Å–∏—è –∑–∞—Ä–µ–≥–∏—Å—Ç—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π —É–≤–µ–ª–∏—á–∏–ª–∞—Å—å –∫–∞–∫ –º–∏–Ω–∏–º—É–º –Ω–∞ 3%.

```python
from statsmodels.stats.proportion import proportions_ztest

# –†–∞–∑–º–µ—Ä—ã –≤—ã–±–æ—Ä–æ–∫ –ê –∏ –í
n_a = participants_events[participants_events['group']=='A']['user_id'].nunique() 
n_b = participants_events[participants_events['group']=='B']['user_id'].nunique()


# –ö–æ–ª-–≤–æ —É—Å–ø–µ—Ö–æ–≤ (–ø–æ–∫—É–ø–æ–∫) –≤ –≥—Ä—É–ø–ø–µ –ê
m_a = participants_events[(participants_events['event_name'] == 'purchase')&(participants_events['group']=='A')]['user_id'].nunique()

# –ö–æ–ª-–≤–æ —É—Å–ø–µ—Ö–æ–≤ (–ø–æ–∫—É–ø–æ–∫) –≤ –≥—Ä—É–ø–ø–µ –í
m_b = participants_events[(participants_events['event_name'] == 'purchase')&(participants_events['group']=='B')]['user_id'].nunique()

p_a = m_a/n_a

p_b = m_b/n_b

print(f'n_a = {n_a}, n_b = {n_b}')
print(f'm_a = {m_a}, m_b = {m_b}')
print(f'p_a = {p_a}, p_b = {p_b}')

alpha = 0.05

stat_ztest, p_value_ztest = proportions_ztest(
    [m_a, m_b], 
    [n_a, n_b], 
    alternative = 'smaller'
    )

print(f'p_value = {p_value_ztest}')

if p_value_ztest<alpha:
    print('–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –æ —Ç–æ–º, —á—Ç–æ –≤ —Ç–µ—á–µ–Ω–∏–µ —Å–µ–º–∏ –¥–Ω–µ–π –ø–æ—Å–ª–µ —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏ –≤ —Å–∏—Å—Ç–µ–º–µ –∫–æ–Ω–≤–µ—Ä—Å–∏—è –∑–∞—Ä–µ–≥–∏—Å—Ç—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –Ω–µ –∏–∑–º–µ–Ω–∏–ª–∞—Å—å –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è. –í —Ç–µ—á–µ–Ω–∏–µ —Å–µ–º–∏ –¥–Ω–µ–π –ø–æ—Å–ª–µ —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏ –≤ —Å–∏—Å—Ç–µ–º–µ –∫–æ–Ω–≤–µ—Ä—Å–∏—è –∑–∞—Ä–µ–≥–∏—Å—Ç—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏–∑–º–µ–Ω–∏–ª–∞—Å—å –º–∏–Ω–∏–º—É–º –Ω–∞ 3 %.')
else:
    print('–ù—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, —á—Ç–æ –≤ —Ç–µ—á–µ–Ω–∏–µ —Å–µ–º–∏ –¥–Ω–µ–π –ø–æ—Å–ª–µ —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏ –≤ —Å–∏—Å—Ç–µ–º–µ –∫–æ–Ω–≤–µ—Ä—Å–∏—è –∑–∞—Ä–µ–≥–∏—Å—Ç—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –Ω–µ –∏–∑–º–µ–Ω–∏–ª–∞—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å.')
```
